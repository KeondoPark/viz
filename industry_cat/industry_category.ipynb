## 해당 label 다 넣기

```python
list_dummy = [Software, Financial_Services, E_Commerce, Health_Care, AI, Security, Manufacturing, Education, Transportation, Food]
data['ind_list'] = ''

for i in range(1000):
    flag = False
    ind = str(data['Industries'][i]).split(',')
    for tmp in range(len(ind)):
        ind[tmp] = ind[tmp].strip()
    for j in range(len(list_dummy)):
        if( set(ind) & set(list_dummy[j])):
            data['ind_list'][i] += str(j+1)+','
            flag= True
    data['ind_list'][i] = data['ind_list'][i][:-1]
    if(flag==False):
        data['ind_list'][i]='11'
            
```

```python
a=''
for i in data['ind_list']:
    a+=str(i)+','
aa=a.split(',')
pd.Series(aa).value_counts()
```

```python
data.to_csv('industy_cat_ver1.csv')
```
